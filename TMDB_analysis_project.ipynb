```python
titles=db_m.nlargest(5,["budget"])
titles2=titles["original_title"]
budget=titles["budget"]
fig=plt.figure(figsize=(25,10))
ax=fig.add_subplot(111)
ax.set_title("the top 5 movies")
ax.set_xlabel("the movies title")
ax.set_ylabel("the budget")
plt.bar(titles2,budget)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm 
fig=plt.figure(figsize=(20,20))
ax=fig.gca(projection="3d")
x=np.arange(-10,10,0.4)
y=np.arange(-10,10,0.4)
x, y = np.meshgrid(x,y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)
surf=ax.plot_surface(x,y,z, cmap=cm.Spectral)
fig.colorbar(surf,shrink=0.5,aspect=5)
plt.show()
```
